```python
import requests
from bs4 import BeautifulSoup

scraped_Titles = []  

for x in range(2, 10):
    mylink = f'https://www.buyrentkenya.com/flats-apartments-for-rent/nairobi/kileleshwa?page={x}'

    response = requests.get(mylink)
    result = BeautifulSoup(response.content, "html.parser")
    result

    ALL_LINKS = result.find_all(class_='text-black no-underline')

    links = [l.get('href') for l in ALL_LINKS]

    linksS = [f'https://www.buyrentkenya.com{l}' for l in links]

    new_links = []
    for link in linksS:
        new_linkss = link.split("\n")
        new_links.append(new_linkss)
    new_links

    second_links = []

    for x in range(0, len(new_links), 2):
        second_links.append(new_links[x])

    total_links = []
    for x in second_links:
        total_links.append(x)

    link_list = []
    for x in total_links:
        link_s = x[0]
        link_list.append(link_s)
    link_list

    for link in link_list:
        response = requests.get(link)
        myresult = BeautifulSoup(response.content, "html.parser")

        number_titles = myresult.find(class_="text-lg font-semibold md:text-2xl")  
        
        if number_titles:
            titles = number_titles.get_text(strip=True)  
            scraped_Titles.append(titles)  
        else:
            scraped_Titles.append("Title not found for this link")  
```

```python
import requests
from bs4 import BeautifulSoup

scraped_Bathrooms = []

for x in range(2, 10):
    mylink = f'https://www.buyrentkenya.com/flats-apartments-for-rent/nairobi/kileleshwa?page={x}'

    response = requests.get(mylink)
    result = BeautifulSoup(response.content, "html.parser")
    result

    ALL_LINKS = result.find_all(class_='text-black no-underline')

    links = [l.get('href') for l in ALL_LINKS]

    linksS = [f'https://www.buyrentkenya.com{l}' for l in links]

    new_links = []
    for link in linksS:
        new_linkss = link.split("\n")
        new_links.append(new_linkss)
    new_links

    second_links = []

    for x in range(0, len(new_links), 2):
        second_links.append(new_links[x])

    total_links = []
    for x in second_links:
        total_links.append(x)

    link_list = []
    for x in total_links:
        link_s = x[0]
        link_list.append(link_s)
    link_list

    for link in link_list:
        response = requests.get(link)
        myresult = BeautifulSoup(response.content, "html.parser")

        number_bathrooms = myresult.find(attrs={"aria-label": "bathrooms"})

        if number_bathrooms:
            bathrooms = number_bathrooms.get_text(strip=True)
            scraped_Bathrooms.append(bathrooms)
        else:
            scraped_Bathrooms.append("0")
```

```python
import requests
from bs4 import BeautifulSoup

scraped_Prices = []

for x in range(2, 10):
    mylink = f'https://www.buyrentkenya.com/flats-apartments-for-rent/nairobi/kileleshwa?page={x}'

    response = requests.get(mylink)
    result = BeautifulSoup(response.content, "html.parser")
    result

    ALL_LINKS = result.find_all(class_='text-black no-underline')

    links = [l.get('href') for l in ALL_LINKS]

    linksS = [f'https://www.buyrentkenya.com{l}' for l in links]

    new_links = []
    for link in linksS:
        new_linkss = link.split("\n")
        new_links.append(new_linkss)
    new_links

    second_links = []

    for x in range(0, len(new_links), 2):
        second_links.append(new_links[x])

    total_links = []
    for x in second_links:
        total_links.append(x)

    link_list = []
    for x in total_links:
        link_s = x[0]
        link_list.append(link_s)
    link_list

    for link in link_list:
        response = requests.get(link)
        myresult = BeautifulSoup(response.content, "html.parser")

        number_prices = myresult.find(class_="text-xl leading-7 font-semibold md:text-xxl md:font-extrabold block")

        if number_prices:
            prices = number_prices.get_text(strip=True)
            scraped_Prices.append(prices)
        else:
            scraped_Prices.append("Title not found for this link")
```

```python
import pandas as pd

# Read the CSV file
Kileleshwa = pd.read_csv('Rent_kILELESHWA_Data.csv')

# Extract data from the Kileleshwa DataFrame
price = Kileleshwa["Prices"]
mypricelist = list(price)

# Clean the prices data
cleaned_prices = [int(price.replace('KSh ', '').replace(',', '')) if 'KSh' in price else None for price in mypricelist]
prices = [price for price in cleaned_prices]

title = Kileleshwa["Title"]
mytitlelist = list(title)

bedrooms = Kileleshwa["Bedrooms"]
mybedroomslist = list(bedrooms)

bathrooms = Kileleshwa["Bathrooms"]
mybathroomslist = list(bathrooms)

# Create a DataFrame for Kileleshwa
KileleshwaDataframe = pd.DataFrame({
    "Title": mytitlelist,
    "Bedrooms": mybedroomslist,
    "Bathrooms": mybathroomslist,
    "Prices": prices
})

# Create a DataFrame for Kilimani (assuming it's a separate DataFrame)
KilimaniDataframe = pd.DataFrame({
    "Title": mytitlelist,
    "Bedrooms": mybedroomslist,
    "Bathrooms": mybathroomslist,
    "Prices": prices
})

# Find indices where 'Bedrooms' is not found for this link
indices_to_remove = KilimaniDataframe[KilimaniDataframe['Bedrooms'] == 'Title not found for this link'].index

# Drop rows with the specified indices
KileleshwaDataframe = KilimaniDataframe.drop(indices_to_remove)

# Reset the index of the DataFrame
KileleshwaDataframe = KileleshwaDataframe.reset_index(drop=True)

# Display the resulting DataFrame
KileleshwaDataframe
```
