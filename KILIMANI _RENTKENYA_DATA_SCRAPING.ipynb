```python
import requests
from bs4 import BeautifulSoup

scraped_Prices = []

for x in range(2, 10):
    mylink = f'https://www.buyrentkenya.com/flats-apartments-for-rent/nairobi/kilimani?page={x}'

    response = requests.get(mylink)
    result = BeautifulSoup(response.content, "html.parser")
    result

    ALL_LINKS = result.find_all(class_='text-black no-underline')

    links = [l.get('href') for l in ALL_LINKS]

    linksS = [f'https://www.buyrentkenya.com{l}' for l in links]

    new_links = []
    for link in linksS:
        new_linkss = link.split("\n")
        new_links.append(new_linkss)
    new_links

    second_links = []

    for x in range(0, len(new_links), 2):
        second_links.append(new_links[x])

    total_links = []
    for x in second_links:
        total_links.append(x)

    link_list = []
    for x in total_links:
        link_s = x[0]
        link_list.append(link_s)
    link_list

    for link in link_list:
        response = requests.get(link)
        myresult = BeautifulSoup(response.content, "html.parser")

        number_prices = myresult.find(class_="text-xl leading-7 font-semibold md:text-xxl md:font-extrabold block")

        if number_prices:
            prices = number_prices.get_text(strip=True)
            scraped_Prices.append(prices)
        else:
            scraped_Prices.append("Title not found for this link")
```

```python
import requests
from bs4 import BeautifulSoup

scraped_Bathrooms = []

for x in range(2, 10):
    mylink = f'https://www.buyrentkenya.com/flats-apartments-for-rent/nairobi/kilimani?page={x}'

    response = requests.get(mylink)
    result = BeautifulSoup(response.content, "html.parser")
    result

    ALL_LINKS = result.find_all(class_='text-black no-underline')

    links = [l.get('href') for l in ALL_LINKS]

    linksS = [f'https://www.buyrentkenya.com{l}' for l in links]

    new_links = []
    for link in linksS:
        new_linkss = link.split("\n")
        new_links.append(new_linkss)
    new_links

    second_links = []

    for x in range(0, len(new_links), 2):
        second_links.append(new_links[x])

    total_links = []
    for x in second_links:
        total_links.append(x)

    link_list = []
    for x in total_links:
        link_s = x[0]
        link_list.append(link_s)
    link_list

    for link in link_list:
        response = requests.get(link)
        myresult = BeautifulSoup(response.content, "html.parser")

        number_bathrooms = myresult.find(attrs={"aria-label": "bathrooms"})

        if number_bathrooms:
            bathrooms = number_bathrooms.get_text(strip=True)
            scraped_Bathrooms.append(bathrooms)
        else:
            scraped_Bathrooms.append("0")
```
