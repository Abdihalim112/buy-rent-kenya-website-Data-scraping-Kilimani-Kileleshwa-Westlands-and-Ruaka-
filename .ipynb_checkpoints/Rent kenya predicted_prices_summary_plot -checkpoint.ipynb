```python
from sklearn.linear_model import LinearRegression
```

```python
price = Kilimani["Prices"]
mypricelist = list(price )

mypricelist


cleaned_prices = [int(price.replace('KSh ', '').replace(',', '')) if 'KSh' in price else None for price in mypricelist]
prices = []
for price in cleaned_prices:
    prices.append(price)


prices

title = Kilimani["Title"]
mytitlelist = list(title)

bedrooms = Kilimani["Bedrooms"]
mybedroomslist = list(bedrooms)

Bathrooms = Kilimani["Bathrooms"]
mybathroomslist = list(Bathrooms)

import pandas as pd
KilimaniDataframe = pd.DataFrame({"Title" : mytitlelist,"Bedrooms" : mybedroomslist,"Bathrooms" : mybathroomslist , "Prices" :prices })

KilimaniDataframe

import pandas as pd

KilimaniDataframe = pd.DataFrame({
    "Title": mytitlelist,
    "Bedrooms": mybedroomslist,
    "Bathrooms": mybathroomslist,
    "Prices": prices
})

```

```python
indices_to_remove = [18, 98]


KilimaniDataframe = KilimaniDataframe.drop(indices_to_remove)


KilimaniDataframe = KilimaniDataframe.reset_index(drop=True)


KilimaniDataframe


```

```python
features = ['Bedrooms','Bathrooms']
target = 'Prices'


KilimaniDataframe = KilimaniDataframe.dropna(subset=[target])


x = KilimaniDataframe[features]
y = KilimaniDataframe[target]


model = LinearRegression()
model.fit(x, y)
```

```python
new_data_kilimani = pd.DataFrame([[2,1]], columns = features)

new_data_kilimani

predicted_price_kilimani = model.predict(new_data_kilimani)

print("Predicted Price kilimani" ,predicted_price_kilimani)
```

```python
import pandas as pd

# Read the CSV file
Kileleshwa = pd.read_csv('Rent_kILELESHWA_Data.csv')

# Extract data from the Kileleshwa DataFrame
price = Kileleshwa["Prices"]
mypricelist = list(price)

# Clean the prices data
cleaned_prices = [int(price.replace('KSh ', '').replace(',', '')) if 'KSh' in price else None for price in mypricelist]
prices = [price for price in cleaned_prices]

title = Kileleshwa["Title"]
mytitlelist = list(title)

bedrooms = Kileleshwa["Bedrooms"]
mybedroomslist = list(bedrooms)

bathrooms = Kileleshwa["Bathrooms"]
mybathroomslist = list(bathrooms)

# Create a DataFrame for Kileleshwa
KileleshwaDataframe = pd.DataFrame({
    "Title": mytitlelist,
    "Bedrooms": mybedroomslist,
    "Bathrooms": mybathroomslist,
    "Prices": prices
})

# Create a DataFrame for Kilimani (assuming it's a separate DataFrame)
KilimaniDataframe = pd.DataFrame({
    "Title": mytitlelist,
    "Bedrooms": mybedroomslist,
    "Bathrooms": mybathroomslist,
    "Prices": prices
})

# Find indices where 'Bedrooms' is not found for this link
indices_to_remove = KilimaniDataframe[KilimaniDataframe['Bedrooms'] == 'Title not found for this link'].index

# Drop rows with the specified indices
KileleshwaDataframe = KilimaniDataframe.drop(indices_to_remove)

# Reset the index of the DataFrame
KileleshwaDataframe = KileleshwaDataframe.reset_index(drop=True)

# Display the resulting DataFrame
KileleshwaDataframe
```
